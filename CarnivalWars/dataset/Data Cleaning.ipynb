```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
train_data.shape,test_data.shape
```

```python
train_data['charges_1']=train_data['charges_1'].fillna(train_data['charges_1'].median())
train_data['charges_2 (%)']=train_data['charges_2 (%)'].fillna(train_data['charges_2 (%)'].median())
train_data['Minimum_price']=train_data['Minimum_price'].fillna(train_data['Minimum_price'].median())
train_data['Maximum_price']=train_data['Maximum_price'].fillna(train_data['Maximum_price'].median())
train_data['Selling_Price']=train_data['Selling_Price'].fillna(train_data['Selling_Price'].median())

```

```python
import scipy.stats as stat
from matplotlib import pylab
from pylab import *
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1) #1 row 2 columns
    df[feature].hist()
    plt.subplot(1,2,2) #1st row 2nd column 2nd index
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(train_data,'Maximum_price')
```

```python
import numpy as np
train_data['Maximum_price_trans']=np.log(train_data['Maximum_price'])
plot_data(train_data,'Maximum_price_trans')
```

```python
import numpy as np
train_data['Min_price_trans']=np.log(train_data['Minimum_price'])
plot_data(train_data,'Min_price_trans')
```

```python
#SQUARE ROOT TRANSFORMATION
train_data['Minprice_square']=train_data.Minimum_price**(1/2)
plot_data(train_data,'Minprice_square')
```

```python
import numpy as np
train_data['charges_1_trans']=np.log(train_data['charges_1'])
plot_data(train_data,'charges_1_trans')
```

```python
#SQUARE ROOT TRANSFORMATION
train_data['charges_1_sq']=train_data.charges_1**(1/2)
plot_data(train_data,'charges_1_sq')
```
